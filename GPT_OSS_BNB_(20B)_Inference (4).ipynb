### Reasoning Effort
The `gpt-oss` models from OpenAI include a feature that allows users to adjust the model's "reasoning effort." This gives you control over the trade-off between the model's performance and its response speed (latency) which by the amount of token the model will use to think.

----

The `gpt-oss` models offer three distinct levels of reasoning effort you can choose from:

* **Low**: Optimized for tasks that need very fast responses and don't require complex, multi-step reasoning.
* **Medium**: A balance between performance and speed.
* **High**: Provides the strongest reasoning performance for tasks that require it, though this results in higher latency.

```python
from transformers import TextStreamer

messages = [
    {"role": "user", "content": "i am batman"},
]
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True,
    return_tensors = "pt",
    return_dict = True,
    reasoning_effort = "high", # **NEW!** Set reasoning effort to low, medium or high
).to(model.device)

_ = model.generate(**inputs, max_new_tokens = 512, streamer = TextStreamer(tokenizer))
```

```python
import torch
from transformers import TextIteratorStreamer
import threading
import gradio as gr

# ‚úÖ assumes you already loaded `model` and `tokenizer`

def generate_response(message, history):
    messages = [{"role": "user", "content": message}]

    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True,
        reasoning_effort="high",   # ‚úÖ force reasoning mode
    ).to(model.device)

    streamer = TextIteratorStreamer(
        tokenizer,
        skip_prompt=False,         # show user + assistant prompt
        skip_special_tokens=False  # don't hide reasoning tags
    )

    thread = threading.Thread(
        target=model.generate,
        kwargs=dict(
            **inputs,
            max_new_tokens=512,
            streamer=streamer,
            do_sample=True,
            temperature=0.7,
        )
    )
    thread.start()

    output_text = ""
    for new_text in streamer:
        output_text += new_text
        yield output_text

# ‚úÖ Gradio app
with gr.Blocks() as demo:
    gr.Markdown("## OSS 20B Chatbot (Reasoning Mode)")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Type your message")
    clear = gr.Button("Clear")

    def user_input(user_message, history):
        return "", history + [[user_message, ""]]

    def bot_response(history):
        user_message = history[-1][0]
        response = generate_response(user_message, history)
        for partial in response:
            history[-1][1] = partial
            yield history

    msg.submit(user_input, [msg, chatbot], [msg, chatbot], queue=False).then(
        bot_response, chatbot, chatbot
    )
    clear.click(lambda: None, None, chatbot, queue=False)

demo.launch(share=True)
```

This cell sets up a Gradio web interface to interact with the OSS-20B model in reasoning mode.

Functionality:

Provides a chat interface where users type prompts and receive streamed responses in real time.

Maintains chat history so the model can respond contextually across turns.

Includes a clear button to reset the conversation.

Shares the app publicly with a temporary Gradio link for remote access.

```python
import torch
from transformers import TextIteratorStreamer
import threading
import gradio as gr

# ‚úÖ assumes you already loaded `model` and `tokenizer`

def generate_response(history, role, system_prompt, reasoning_effort, max_new_tokens, temperature):
    # Build messages from chat history
    messages = []
    if system_prompt.strip():
        messages.append({"role": "system", "content": system_prompt})

    # Convert chat history into messages list
    for user_msg, bot_msg in history:
        if user_msg:
            messages.append({"role": "user", "content": user_msg})
        if bot_msg:
            messages.append({"role": "assistant", "content": bot_msg})

    # Add the *latest* input role (user/system/assistant)
    if history and history[-1][0] and not history[-1][1]:
        messages.append({"role": role, "content": history[-1][0]})

    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True,
        reasoning_effort=reasoning_effort,
    ).to(model.device)

    streamer = TextIteratorStreamer(
        tokenizer,
        skip_prompt=False,
        skip_special_tokens=False
    )

    thread = threading.Thread(
        target=model.generate,
        kwargs=dict(
            **inputs,
            max_new_tokens=max_new_tokens,
            streamer=streamer,
            do_sample=True,
            temperature=temperature,
        )
    )
    thread.start()

    output_text = ""
    for new_text in streamer:
        output_text += new_text
        yield output_text

# ‚úÖ Gradio app
with gr.Blocks() as demo:
    gr.Markdown("## OSS 20B Chatbot (Reasoning Mode + Full Chat History)")

    with gr.Row():
        role = gr.Dropdown(["user", "assistant", "system"], value="user", label="Role")
        system_prompt = gr.Textbox(label="System Prompt", placeholder="e.g., You are a helpful assistant.")

    with gr.Row():
        reasoning_effort = gr.Dropdown(["low", "medium", "high"], value="high", label="Reasoning Effort")
        max_new_tokens = gr.Slider(50, 1024, value=512, step=10, label="Max Tokens")
        temperature = gr.Slider(0.1, 1.5, value=0.7, step=0.1, label="Temperature")

    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Type your message")
    clear = gr.Button("Clear")

    def user_input(user_message, history):
        return "", history + [[user_message, ""]]

    def bot_response(history, role, system_prompt, reasoning_effort, max_new_tokens, temperature):
        response = generate_response(
            history, role, system_prompt, reasoning_effort, max_new_tokens, temperature
        )
        for partial in response:
            history[-1][1] = partial
            yield history

    msg.submit(
        user_input, [msg, chatbot], [msg, chatbot], queue=False
    ).then(
        bot_response, [chatbot, role, system_prompt, reasoning_effort, max_new_tokens, temperature], chatbot
    )

    clear.click(lambda: None, None, chatbot, queue=False)

demo.launch(share=True)
```

This cell sets up a Gradio-powered chatbot with full control over conversation roles, system prompts, reasoning settings, and generation parameters.

Functionality:

Interactive chat interface with chat history tracking.

Users can dynamically set the role of the next input (user, assistant, or system).

Allows custom system prompts for role alignment or behavior steering.

Adjustable generation settings from the UI:

Reasoning effort (low, medium, high)

Max new tokens (50‚Äì1024, default 512)

Temperature (0.1‚Äì1.5, default 0.7)

Includes a clear button to reset the conversation thread.

App is served locally and shared via a Gradio public link.

Technical details:

Chat history is serialized into a list of role-based messages (system, user, assistant) and passed through Hugging Face‚Äôs tokenizer.apply_chat_template.

reasoning_effort is explicitly controlled at runtime, passed into the tokenizer.

Model generation is run in a separate thread for non-blocking streaming.

TextIteratorStreamer is used for token-by-token output streaming with full prompt and reasoning visibility (skip_prompt=False, skip_special_tokens=False).

Parameters (role, system_prompt, reasoning_effort, max_new_tokens, temperature) are exposed as client-side controls, enabling experimentation without modifying backend code.

## Why Unsloth?

Using the base version of large language models like GPT-OSS 20B often requires significant computational resources (GPU memory, processing power) that may not be readily available on free Google Colab instances. Unsloth is a library designed to optimize the process of loading and fine-tuning large language models, making it possible to work with these models efficiently on more constrained hardware, such as the free Tesla T4 GPUs provided by Colab.

Unsloth achieves this through various techniques, including:

- **Optimized Memory Usage:** Reducing the memory footprint of the model.
- **Faster Loading:** Speeding up the process of loading the model weights.
- **Integration with Quantization:** Supporting techniques like 4-bit quantization to further reduce memory requirements.

In this notebook, we use Unsloth to load a pre-quantized version of the GPT-OSS 20B model (`unsloth/gpt-oss-20b-unsloth-bnb-4bit`), which is specifically optimized for use with Unsloth and BitsAndBytes 4-bit quantization.

## Reasoning Effort

The `gpt-oss` models from OpenAI introduce a "reasoning effort" feature. This allows users to control the trade-off between the model's response speed (latency) and its performance on tasks requiring complex reasoning. By adjusting the reasoning effort, you can influence how much "thinking" the model does before generating a response, which is often reflected in the amount of internal "analysis" tokens it generates.

The notebook demonstrates the three levels of reasoning effort:

- **Low:** Prioritizes speed for simple tasks.
- **Medium:** Balances performance and speed.
- **High:** Prioritizes strong reasoning for complex tasks, potentially with higher latency.

The notebook shows examples of using the model with "medium" and "high" reasoning effort.
